Data PreProcessing

```python
# One hot encoding
# Label Encode
# positive -> 1
# negative -> 2
df.replace({"sentiment":{"positive": 1, "negative": 2}}, inplace=True)
df.head()
```

```python
def data_processing(text):
  text = text.lower()
  text = re.sub('<br />', '', text)            # removing br tags
  text = re.sub(r"https\S+|www\S+https\S+", '', text, flags=re.MULTILINE)   # removing the URLs
  text = re.sub(r'\@w+|\#', '', text)   # removing the hastags
  text = re.sub(r'[^\w\s]', '', text)   # removing the punctuations
  text_tokens = word_tokenize(text)
  filtered_text = [w for w in text_tokens if not w in stop_words]
  return " ".join(filtered_text)
```

```python
df.review = df['review'].apply(data_processing)
```

```python
duplicated_count = df.duplicated().sum()
print("Number of duplicated reviews: ", duplicated_count)
```

```python
# removing the duplicate reviews
df = df.drop_duplicates('review')
```

Data Extraction

```python
stremer = PorterStemmer()
def stemming(data):
  text = [stremer.stem(word) for word in data]
  return data
```

```python
df.review = df['review'].apply(lambda x: stemming(x))
```

```python
df['word count'] = df['review'].apply(no_of_words)
df.head()
```

```python
X = df['review']
Y = df['sentiment']
```

```python
vect = TfidfVectorizer()
X = vect.fit_transform(df['review'])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state=42)
```

```python
print("Size of X_train:", (X_train.shape))
print("Size of Y_train:", (Y_train.shape))
print("Size of X_test:", (X_test.shape))
print("Size of Y_test:", (Y_test.shape))
```

Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
logreg_pred = logreg.predict(X_test)
logreg_acc = accuracy_score(logreg_pred, Y_test)
print("Test accuracy: {:.2f}%".format(logreg_acc*100))
```

```python
mnb = MultinomialNB()
mnb.fit(X_train, Y_train)
mnb_pred = mnb.predict(X_test)
mnb_acc = accuracy_score(mnb_pred, Y_test)
print("Test accuracy: {:.2f}%".format(mnb_acc*100))
```

```python
svc = LinearSVC()
svc.fit(X_train, Y_train)
svc_pred = svc.predict(X_test)
svc_acc = accuracy_score(svc_pred, Y_test)
print("Test accuracy: {:.2f}%".format(svc_acc*100))
```

Evaluation

```python
print(confusion_matrix(Y_test, logreg_pred))
print("\n")
print(classification_report(Y_test, logreg_pred))
```

```python
print(confusion_matrix(Y_test, mnb_pred))
print("\n")
print(classification_report(Y_test, mnb_pred))
```

```python
print(confusion_matrix(Y_test, svc_pred))
print("\n")
print(classification_report(Y_test, svc_pred))
```
